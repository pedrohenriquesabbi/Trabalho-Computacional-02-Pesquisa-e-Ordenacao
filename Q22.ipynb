```python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:

    def __init__(self):
        self.head = None

    def append(self, data):

        new_node = Node(data)


        if not self.head:
            self.head = new_node
            return


        last = self.head
        while last.next:
            last = last.next


        last.next = new_node

    def print_list(self):

        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    def get_tail(self):

        if not self.head:
            return None

        current = self.head
        while current.next:
            current = current.next
        return current

    def partition(self, head, end, new_head_ref, new_end_ref):

        pivot = end
        prev = None
        current = head
        tail = pivot

        while current != pivot:
            if current.data < pivot.data:

                if new_head_ref[0] is None:
                    new_head_ref[0] = current

                prev = current
                current = current.next
            else:

                if prev:
                    prev.next = current.next

                temp = current.next
                current.next = None
                tail.next = current
                tail = current
                current = temp


        if new_head_ref[0] is None:
            new_head_ref[0] = pivot


        new_end_ref[0] = tail

        return pivot

    def quick_sort_recur(self, head, end):

        if not head or head == end:
            return head

        new_head = [None]
        new_end = [None]


        pivot = self.partition(head, end, new_head, new_end)


        if new_head[0] != pivot:

            temp = new_head[0]
            while temp.next != pivot:
                temp = temp.next
            temp.next = None


            new_head[0] = self.quick_sort_recur(new_head[0], temp)


            temp = self.get_tail_from_node(new_head[0])
            temp.next = pivot


        pivot.next = self.quick_sort_recur(pivot.next, new_end[0])

        return new_head[0]

    def get_tail_from_node(self, node):

        current = node
        while current and current.next:
            current = current.next
        return current

    def quick_sort(self):

        self.head = self.quick_sort_recur(self.head, self.get_tail())


# Demonstração de uso
if __name__ == "__main__":
    print("=== Quick Sort em Lista Encadeada ===")
    ll = LinkedList()
    numeros = [30, 10, 50, 20, 40, 60, 25]

    print("\nAdicionando números à lista:")
    for num in numeros:
        ll.append(num)
        print(f"Adicionado {num}: ", end="")
        ll.print_list()

    print("\nLista antes da ordenação:")
    ll.print_list()

    ll.quick_sort()

    print("\nLista após ordenação:")
    ll.print_list()
```
